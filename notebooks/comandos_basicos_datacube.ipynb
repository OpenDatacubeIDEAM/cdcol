```python
import datacube
#Sin la ruta del datacube.conf funciona
#El app es un identificado
dc = datacube.Datacube(app = 'odc_costa_rica')
#Si está bien la conexión a la base de datos, va mostrar algo de este estilo 
# Datacube<index=Index<db=PostgresDb<engine=Engine(postgresql://ubuntu:***@localhost:5432/datacube)>>>
print(dc)
```

```python
#Esto muestra los productos disponibles del datacube
#Por defecto se crean algunos
#El que nosotros creamos es LS8_OLI_LASRC
dc.list_products()
```

```python
#Esto muestra las bandas disponibles por cada producto
dc.list_measurements()
```

```python
#Esta función nos muestra los datasets ingestados del producto que se pasa como parámetro
#En este caso las de LANDSAT 8
dc.find_datasets(product='LS8_OLI_LASRC')
```

```python
#Este es un plus, usando la librería netCDF4 que permite leer un archivo netcdf(.nv)
from netCDF4 import Dataset
#La función dataset nos permite abrir un archivo .nc para manipularlo. En este caso es una de las ingestadas
nc = Dataset("/home/ubuntu/dc_storage/LS8_OLI_LASRC/LS8_OLI_LASRC_4326_-74_4_20130814145001000000.nc")
#Cuando se imprime dataset, se pueden ver los valores de las variables 
print(nc)
#Muestra las variables disponibles
nc.variables
#Muestra las dimensiones disponibles
nc.dimensions
```

```python
#La función load del datacube permite hacer una consulta sobre un producto, con una latitud, longitud y tiempo
xarr = dc.load(product='LS8_OLI_LASRC', latitude=(3, 5), longitude=(-69, -68), time=('2013-01-01','2013-12-31'))
print(xarr.data_vars)
```
